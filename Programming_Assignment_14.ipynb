<b>Question 1:
Define a class with a generator which can iterate the numbers, which are divisible by
7, between a given range 0 and n.

```python
def div_by_7(n):
    for i in range(n + 1):
        if i % 7 == 0:
            yield i
            
div_by_7_gen = div_by_7(70)
for i in div_by_7_gen:
    print(i)
```

<b>Question 5:
Please write a program to compress and decompress the string &quot;hello world!hello
world!hello world!hello world!&quot;.

```python
def compress(s):
    s = list(s + ' ')
    count = 1
    comp = ''
    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            if s[i] == s[j]:
                count += 1
                break
            else:
                comp += s[i]
                if count != 1:
                    comp += str(count)
                count = 1
                break
    return comp
```

```python
compress('hello world!hello world!hello world!hello world!')
```

```python
compress('aabbc ccddefggghhijj jjklmnnni')
```

```python
compress('aabbbb')
```

```python
def decompress(s):
    s = list(s + ' ')
    comp = ''
    for i in range(len(s)):
        for j in range(i + 1, len(s)):
            if s[i].isdigit() == False:
                if s[j].isdigit():
                    comp += s[i] * int(s[j])
                    break
                else:
                        comp += s[i]
                        break
    return comp
```

```python
decompress('a3b2c1d5eef5ggh1')
```

```python
decompress('a3b2c1d5e ef5ggh1')
```

```python
assert decompress(compress('hello world!hello world!hello world!hello world!')) ==\
'hello world!hello world!hello world!hello world!'
```

<b>Question 6:
Please write a binary search function which searches an item in a sorted list. The
function should return the index of element to be searched in the list.

```python
# referred from https://www.geeksforgeeks.org/python-program-for-binary-search/
def binary_search(l, x):
    low = 0
    high = len(l) - 1
    mid = 0
    while low <= high:
        mid = (high + low) // 2
        if l[mid] < x:
            low = mid + 1
        elif l[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
```

```python
binary_search([1,34,55,66,77],66)
```
